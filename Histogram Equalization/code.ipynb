# Reading image

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
# opening image
img = Image.open('image.png')
arr = np.array(img,dtype=np.uint8)
```

```python
plt.imshow(arr)
```

# Converting image to grayscale

```python
grayscale=np.ones((arr.shape[0],arr.shape[1]),dtype=np.uint8)
```

```python
# averaging rgb to achieve grayscale
for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        tmp=(float(arr[i][j][0])+float(arr[i][j][1])+float(arr[i][j][2]))/3
        grayscale[i][j]=int(tmp)
```

```python
plt.imshow(grayscale,cmap="gray",vmin=0,vmax=255)
```

# Histogram

```python
colors_repetition={}
for i in range(grayscale.shape[0]):
    for j in range(grayscale.shape[1]):
        current_color=grayscale[i][j]
        if current_color not in colors_repetition.keys():
            colors_repetition[current_color]=1
        else:
            colors_repetition[current_color]=colors_repetition[current_color]+1
```

```python
bins=[]
hist_array=[]
for i in range(255):
    bins.append(i)
    if i not in colors_repetition:
        hist_array.append(0)
    else:
        hist_array.append(colors_repetition[i])
```

# plotting histogram:

```python
plt.hist(grayscale.ravel(),bins=bins,density=True)
```

```python
cumulative_arr=[]
tmp=0
for i in range(255):
    if i in colors_repetition.keys():
        tmp+=colors_repetition[i]
    cumulative_arr.append(tmp)
```

```python
new_colors={}
for i in range(255):
    new_colors[i]=round((255-1)*cumulative_arr[i]/(grayscale.shape[0]*grayscale.shape[1]))
```

```python
final_img=np.ones((grayscale.shape[0],grayscale.shape[1]),dtype=np.uint8)
```

```python
for i in range(grayscale.shape[0]):
    for j in range(grayscale.shape[1]):
        tmp=new_colors[grayscale[i][j]]
        final_img[i][j]=int(tmp)
```

```python
plt.imshow(final_img,cmap="gray",vmin=0,vmax=255)
```

```python
final_pic=Image.fromarray(final_img)
final_pic.save("output.png")
```

# Histogram of output:

```python
plt.hist(final_img.ravel(),bins=bins,density=True)
```

# Cumulative plot for input image :

```python
plt.hist(grayscale.ravel(),bins=bins,density=True,cumulative=True)
```

# Cumulative plot for output image :

```python
plt.hist(final_img.ravel(),bins=bins,density=True,cumulative=True)
```
